## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes
import pandas as pd
import numpy as np

baltimore_df = pd.read_csv("dataset/baltimore_9-24-2016_9-30-2017.csv")
boston_df    = pd.read_csv("dataset/boston_9-24-2016_9-30-2017.csv")
newyork_df   = pd.read_csv("dataset/new-york_9-24-2016_9-30-2017.csv")
philly_df    = pd.read_csv("dataset/philadelphia_9-24-2016_9-30-2017.csv")
```

## Clean Your Data

In the last lesson, we cleaned the data related to San Francisco. Pull up your notebook from the last lesson and use it as a reference to clean up these new dataframes.

```python
# Clean your data here!
# BAltimore
print('Baltimore')

#
baltimore_df.shape

# Count missing values per column
baltimore_df.isnull().sum()

# Percentage of missing values per column
# baltimore_df.isnull().mean() * 100

# Drop rows with any missing values
baltimore_df = baltimore_df.drop(["Grade", "Type", "Environment", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode"], axis=1)

# Count missing values per column
baltimore_df.isnull().sum()

```

```python
# Clean your data here!
# Boston
print ('Boston')
# Count missing values per column
boston_df.isnull().sum()

# Percentage of missing values per column
boston_df.isnull().mean() * 100

# Drop rows with any missing values
boston_df = boston_df.drop(["Grade", "Type", "Environment", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode"], axis=1) 

# Count missing values per column
boston_df.isnull().sum()
```

```python
# Clean your data here!
# New York
print('New York')

# Count missing values per column
newyork_df.isnull().sum()

# Percentage of missing values per column
newyork_df.isnull().mean() * 100

# Drop rows with any missing values
newyork_df = newyork_df.drop(["Grade", "Type", "Environment", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode"], axis=1) 

# Count missing values per column
newyork_df.isnull().sum()
```

```python
# Clean your data here!
# Phily
print('Philly')

# Count missing values per column
philly_df.isnull().sum()

philly_df.shape

# Percentage of missing values per column
philly_df.isnull().mean() * 100

# Drop rows with any missing values
philly_df = philly_df.drop(["Grade", "Type", "Environment", "Quality", "Condition", "Appearance", "Storage", "Crop", "Trans Mode"], axis=1) 

# Count missing values per column
philly_df.isnull().sum()

   
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
northeast_df=pd.concat([baltimore_df,boston_df,newyork_df,philly_df],ignore_index=True)

northeast_df.head()

northeast_df.shape
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of **unit of sale** in the Northeast region? 
2. For each region, what is the average number of pumpkins per variety that came into terminal markets for the year? 

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
grouped_by_units_of_sale = northeast_df.groupby("Unit of Sale").agg(
    {"Low Price": "mean", 
     "High Price": "mean"
})

grouped_by_units_of_sale
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
counts_by_variety = northeast_df.groupby("Variety").size().reset_index(name='count')

counts_by_variety
```
